```python
l1=[[1,7,3],[9,5,6],[12,8,9]]
max=0
check=0
for i in range(len(l1)):
    for j in range(len(l1[i])):
        if l1[i][j]>check:
            check=l1[i][j]
    max+=check
print(max)
```
