```python

list1=[2,4,6,8,3,5,8,9]
mid=len(list1)//2
sum1=0
sum2=0
for i in range(0,mid):
    sum1=sum1+list1[i]
for j in range(mid,len(list1)):
    sum2=sum2+list1[j]
print(sum1," ",sum2)
```

```python
l=[2,4,6,8,3,5,8,9]
mid = len(l) // 2

sum1 = sum(l[:mid])   
sum2 = sum(l[mid:])     

print("First half sum:", sum1)
print("Second half sum:", sum2)
```
