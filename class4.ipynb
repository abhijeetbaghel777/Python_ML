```python
#factor of a number in function
def factor(n):
    for i in range(1,n+1):
        if n%i==0:
            print(i,end=" ")
        return 10
a=int(input("Enter the number:"))
factor(a)
```

```python
#to calculate nCr
def combination(n,r):
    c=n-r
    n1=1
    r1=1
    c1=1
    for i in range(1,n+1):
        n1=n1*i
    for i in range(1,r+1):
        r1=r1*i
    for i in range(1,c+1):
        c1=c1*i
    print("Answer=",n1/(c1*r1))
a=int(input("Enter N:"))
b=int(input("Enter R:"))
combination(a,b)
```
