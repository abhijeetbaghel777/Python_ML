```python
#even and odd factor count of number
a=int(input("Enter the number:"))
even=0
odd=0
for i in range(1,a):
    if a%i==0 and a%2==0:
        even=even+1
    elif a%i==0 and a%2!=0:
        odd=odd+1
print("even count:",even,"\nodd count:",odd)
```
