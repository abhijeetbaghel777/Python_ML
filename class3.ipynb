```python
#even and odd factor count of number
a=int(input("Enter the number:"))
even=0
odd=0
for i in range(1,a):
    if a%i==0 and a%2==0:
        even=even+1
    elif a%i==0 and a%2!=0:
        odd=odd+1
print("even count:",even,"\nodd count:",odd)
```

```python
#digit count
a=int(input("Enter the number"))
count=0
while(a>0):
    a=a//10
    count+=1
print("Total digit=",count)
```

```python
#sum of digit
a=int(input("Enter the number:"))
sum=0
while(a>0):
    sum=sum+a%10
    a=a//10
print("sum=",sum)
```
