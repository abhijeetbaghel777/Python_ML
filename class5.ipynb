```python
n = int(input("Enter size: "))
list1 = []

for i in range(n):
    x = int(input("Enter element: "))
    list1.append(x)

print("Before modifying",list1)
for i in range(len(list1)):
    if list1[i]%2==0:
        list1[i]=list1[i]-5
    else:
        list1[i]=list1[i]+5
print("After modifying ",list1)
```

```python
n = int(input("Enter size: "))
list1 = []


for i in range(n):
    x = int(input())
    list1.append(x)
max=list1[0]
max2=list1[0]
print("The list is as follows->",list1)
for i  in range(len(list1)):
    if list1[i]>max:
        max=list1[i]
for i in range(len(list1)):
    if list1[i]>max2 and list1[i]<max:
        max2=list1[i]
print("largest->",max)
print("second largest->",max2)
```
