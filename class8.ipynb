```python
'''🔹 1. What is a String?

A string is a sequence of characters enclosed in quotes.

s1 = "Hello"
s2 = 'World'
s3 = """Multi-line string"""

✔ Strings are immutable → cannot be changed after creation.

🔹 2. Indexing & Accessing

Each character has an index.

s = "Python"

print(s[0])   # P
print(s[-1])  # n (negative indexing)
🔹 3. Slicing

Extract part of a string.

s = "Python"

print(s[0:4])   # Pyth
print(s[:3])    # Pyt
print(s[2:])    # thon
print(s[::-1])  # nohtyP (reverse)
🔹 4. Immutability (Important ⚠️)

You cannot modify a string directly.

s = "hello"
# s[0] = 'H' ❌ ERROR

✔ Correct way:

s = "hello"
s = "H" + s[1:]
print(s)  # Hello
🔹 5. String Operations
➤ Concatenation (+)
a = "Hello"
b = "World"
print(a + " " + b)  # Hello World
➤ Repetition (*)
print("Hi" * 3)  # HiHiHi
🔹 6. Important String Methods
➤ Case Conversion
s = "python"
print(s.upper())   # PYTHON
print(s.capitalize())  # Python
➤ Searching
s = "hello world"
print(s.find("world"))  # index
print(s.count("l"))     # count
➤ Replace
s = "I like Java"
print(s.replace("Java", "Python"))
➤ Split & Join
s = "a,b,c"
lst = s.split(",")   # ['a','b','c']

print(" ".join(lst))  # a b c
➤ Strip (remove spaces)
s = "  hello  "
print(s.strip())
🔹 7. Checking Functions
s = "123"
print(s.isdigit())   # True
print(s.isalpha())   # False
print(s.isalnum())   # True
🔹 8. Looping Through String
s = "Python"

for ch in s:
    print(ch)
🔹 9. Membership Operator
s = "Python"

print("Py" in s)   # True
print("Java" not in s)  # True
🔹 10. String Formatting (Very Important ⭐)
➤ f-string (Best way)
name = "John"
age = 20

print(f"My name is {name} and I am {age}")
🔹 11. Escape Characters
print("Hello\nWorld")  # new line
print("He said \"Hi\"")
🔹 12. Comparison
print("apple" == "apple")  # True
print("a" < "b")           # True (ASCII comparison)
🔹 13. Common Interview Concepts 🔥
✔ Reverse String
s = "python"
print(s[::-1])
✔ Palindrome Check
s = "madam"
print(s == s[::-1])
✔ Remove Spaces
s = "a b c"
print(s.replace(" ", ""))
🔹 14. Mutable Alternative

Since strings are immutable, use:

list (for editing characters)
bytearray (advanced use)
s = "hello"
lst = list(s)
lst[0] = 'H'
s = "".join(lst)
print(s)  # Hello
🔹 Key Takeaways 🧠
Strings are immutable
Support indexing & slicing
Many built-in methods → powerful
Use f-strings for formatting
Use list if you need modification'''
```
